## Import library and data

```python
import tensorflow as tf
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
```

```python
# load fashion mnist dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
# (train_images,train_labels),(test_images,test_labels) = fashion_mnist.load_data()
```

### Splitting data

```python
(X_train,Y_train),(X_test,Y_test) = fashion_mnist.load_data()
```

### Display Dataset


```python
import matplotlib.pyplot as plt
# number of images to display
n = 17
# show image size
plt.figure(figsize=(20,4))

for i in range(n):
    ax = plt.subplot(2,n,i+1)

    # display original image
    plt.imshow(X_test[i].reshape(28,28))

    # set colormap to grayscale
    plt.gray()

    # hide x_axis and y_axis labels and ticks
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    


plt.show()
plt.close()
```

### Flattening image

```python
print("Previous X_train shape: {}  \nPrevious Y_train shape:{} ".format(X_train.shape,Y_train.shape))

X_train = X_train.reshape(60000,784)
X_test = X_test.reshape(10000,784)
```

### Min_Max Scalling

```python
# convert date image to float
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# normalize the image o and 1: o represent black and 1 for white

X_train /= 255
X_test /= 255
```

### Processing the target variable

```python
classes = 15
Y_train = to_categorical(Y_train,classes)
Y_test = to_categorical(Y_test,classes)

print("New X_train shape: {} \nNew Y_train shape: {}".format(X_train.shape,Y_train.shape))
```

### Setting up hyper parameter

```python
input_size = 784
batch_size = 200
# first hidden leyers of neurons
hidden1 = 400
# second hidden layer of neurons
hidden2 = 20

# number of classes 
classes = 15
# set number of complete passes through training
epochs = 10
```

### Building the FCN model

```python
model = Sequential()
# add fist layer using relu
model.add(Dense(hidden1,input_dim=input_size,activation='relu'))
# add second layer of neurons using relu
model.add(Dense(hidden2,activation="relu"))

# add last layer of neurons using softmax
model.add(Dense(classes,activation='softmax'))

model.compile(loss="categorical_crossentropy",metrics=['accuracy'],optimizer='sgd')

model.summary()
```

## **Parameters Calculations**
**output = relu(dot(W, input) + bias)**


(400*784) + 400 = 314000 = 0.3 million paramters


**parameters for Chat-gpt 4 = 1760000000000 = 1.76 trillion parameters**

### Training the model

```python
from time import time
# record currune time
tic = time()
model.fit(X_train,Y_train,batch_size=batch_size,epochs=epochs,verbose=1)
# record after time for measuring training data
toc = time()

print("Model training took {} sec".format(toc - tic))
```

### Testing the model

```python
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# predict probabilities for the test using training model
y_pred_probs = model.predict(X_test,verbose=0)
y_pred = np.where(y_pred_probs >0.5,1,0)

# calculate and print test accuracy
test_accuracy = accuracy_score(y_pred,Y_test)
print("\nTest Accuracy: {}".format(test_accuracy))
```

```python
# define a mask for selecting a range for indices
mask = range(20,50)
X_valid = X_test[0:20]
actual_labels = Y_test[0:20]

y_pred_probs_valid = model.predict(X_valid)
y_pred_valid = np.where(y_pred_probs_valid > 0.5,1,0)
```
